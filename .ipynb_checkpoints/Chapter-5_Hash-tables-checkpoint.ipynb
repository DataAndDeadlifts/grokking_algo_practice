# 5.5, 5.6, 5.7

```python
from string import ascii_lowercase
from matplotlib import pyplot as plt

def primes():
    current = 1
    while True:
        current += 1
        while True:
            for i in range(2, current // 2 + 1):
                if current % i == 0:
                    current += 1
                    break
            else:
                break
        yield current

prime_gen = primes()
        
def hash_A(string):
    return 1

def hash_B(string):
    return len(string)

def hash_C(string):
    return string[0].lower()

prime_map = {}
for c in ascii_lowercase:
    prime_map[c] = next(prime_gen)

def hash_D(string, hash_size=10):
    str_split = list(string)
    prime_sum = sum([prime_map[a.lower()] if a != " " else 1 for a in str_split])
    return prime_sum % hash_size        

def graph_hash_dist(hashes):
    plt.hist(hashes, bins="fd")
    plt.show()
```

```python
# 5.5
phonebook = ["Esther","Ben","Bob","Dan"]
phone_hashes = []
for func in [hash_A, hash_B, hash_C]:
    phone_hashes.append([func(a) for a in phonebook])
    
phone_hashes.append([hash_D(a, int(len(phonebook)/0.7)) for a in phonebook])
    
for entry in phone_hashes:
    print(entry)
    graph_hash_dist(entry)
```

Looks like hash functions C and D produced the greatest distribution.

C was lucky that so many entries didn't share a first character.

Hash function D really shows the greatest flexibility and distribution for strings of this type. If the hash-length were changed here, the function would probably create a greater distribution.

```python
# 5.6
batteries = ["A","AA","AAA","AAAA"]
battery_hashes = []
for func in [hash_A, hash_B, hash_C]:
    battery_hashes.append([func(a) for a in batteries])

battery_hashes.append([hash_D(a, int(len(batteries)/0.7)) for a in batteries])
    
for entry in battery_hashes:
    print(entry)
    graph_hash_dist(entry)
```

Hash functions B and D were equivalent here, only because conveniently the string length was a differentiator here. Both functions incorporate string length.

```python
# 5.7
titles = ["Maus","Fun Home","Watchmen"]
title_hashes = []
for func in [hash_A, hash_B, hash_C]:
    title_hashes.append([func(a) for a in titles])
    
title_hashes.append([hash_D(a, int(len(titles)/0.7)) for a in titles])

    
for entry in title_hashes:
    print(entry)
    graph_hash_dist(entry)
```

C was better here, but only superficially. I would still lean on Hash function D because future titles are likely to share the first character of the title and cause collisions.
